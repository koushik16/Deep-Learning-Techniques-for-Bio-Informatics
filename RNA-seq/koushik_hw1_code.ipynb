```python
import scanpy as sc
import anndata as an
from sklearn.decomposition import PCA
```

```python
from scipy import io
import pandas as pd
input_dir = ""

matrix = io.mmread(input_dir + 'mat.mtx')

peak_list = []
with open(input_dir + "genes.tsv", "r") as filestream:
    for line in filestream:
        peak_list.append(line[:-1])

cell_list = []
with open(input_dir + "barcodes.tsv", "r") as filestream:
    for line in filestream:
        currentline = line.split(",")
        cell_list.append(currentline[0][:-1])
        
df = pd.DataFrame(matrix.toarray(), index=peak_list, columns=cell_list)
```

```python
adata_df = ada.to_df()
display(adata_df.head())
print(adata_df.shape)
```

```python
n_comp = 1997
pca_object = PCA(n_components=n_comp, random_state=500)
pca_object.fit(adata_df)
```

```python
vc=pca_object.explained_variance_ratio_.cumsum()
```

```python
print(vc)
```

```python
ada_pca=pca_object.transform(adata_df)
```

```python
adapca = an.AnnData(ada_pca)
```

```python
ada_proj = pca_object.inverse_transform(ada_pca)
```

```python
adaproj = an.AnnData(ada_proj)
```

```python
adaproj.var_names = ll
```
